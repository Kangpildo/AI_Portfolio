```python

import os
import cv2
import matplotlib.pyplot as plt

# ê²½ë¡œ ì„¤ì • (Step 2ì˜ ê²°ê³¼ ê²½ë¡œ)
save_dir = r"D:\Virtual_Fahion\GraduationThesis2\saveModels"
run_name = "baseline_yolov8n"

def analyze_results(save_dir, run_name):
    print(f"\nğŸ” [Step 3] Analyzing Results for: {run_name}")
    
    base_path = os.path.join(save_dir, run_name)
    results_png = os.path.join(base_path, "results.png")
    confusion_matrix_png = os.path.join(base_path, "confusion_matrix.png")
    
    # ì‹œê°í™” ì¤€ë¹„
    fig, axes = plt.subplots(1, 2, figsize=(24, 10))
    fig.suptitle(f"Analysis Report: {run_name}", fontsize=20, weight='bold')
    
    # 1. í•™ìŠµ ê³¡ì„ 
    if os.path.exists(results_png):
        img = cv2.imread(results_png)
        axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axes[0].set_title("Training Curves (Loss & mAP)", fontsize=14)
        axes[0].axis('off')
    else:
        axes[0].text(0.5, 0.5, "results.png not found", ha='center')
        
    # 2. í˜¼ë™ í–‰ë ¬
    if os.path.exists(confusion_matrix_png):
        img = cv2.imread(confusion_matrix_png)
        axes[1].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axes[1].set_title("Confusion Matrix", fontsize=14)
        axes[1].axis('off')
    else:
        axes[1].text(0.5, 0.5, "confusion_matrix.png not found", ha='center')
        
    plt.tight_layout()
    plt.show()
    print("âœ… Visualization generated.")

# ì‹¤í–‰
if __name__ == "__main__":
    analyze_results(save_dir, run_name)
```
