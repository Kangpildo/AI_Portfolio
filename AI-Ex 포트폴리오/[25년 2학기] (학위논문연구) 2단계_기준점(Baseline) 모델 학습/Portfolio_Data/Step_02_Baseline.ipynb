```python

import os
from ultralytics import YOLO

# Step 1ì˜ í™˜ê²½ ë³€ìˆ˜ë¥¼ ìˆ˜ë™ìœ¼ë¡œ ì„¤ì •í•˜ê±°ë‚˜, ì´ì „ ë‹¨ê³„ ì½”ë“œë¥¼ ì‹¤í–‰í•´ì•¼ í•©ë‹ˆë‹¤.
# ì—¬ê¸°ì„œëŠ” ë…ë¦½ ì‹¤í–‰ì„ ìœ„í•´ ë‹¤ì‹œ ì •ì˜í•©ë‹ˆë‹¤.
base_path = r"D:\Virtual_Fahion\GraduationThesis2"
data_yaml = os.path.join(base_path, "data", "data.yaml")
save_dir = os.path.join(base_path, "saveModels")
device = '0' # GPU ì‚¬ìš© ê°€ì •

def run_baseline_training(model_name, device, data_path, save_path):
    print(f"\nğŸš€ [Step 2] Starting Baseline Training: {model_name}")
    
    # 1. ëª¨ë¸ ì´ˆê¸°í™” (Pre-trained weights ì‚¬ìš©)
    model = YOLO(f"{model_name}.pt")
    
    # 2. í•™ìŠµ ì‹¤í–‰
    results = model.train(
        data=data_path,
        epochs=30,       # ì´ˆê¸° ì‹¤í—˜ìš© 30 epoch
        batch=16,        
        imgsz=640,       
        device=device,
        project=save_path,
        name=f"baseline_{model_name}", 
        exist_ok=True,   
        optimizer='Adam', 
        verbose=True
    )
    
    print(f"âœ… Baseline Training Completed.")
    print(f"ğŸ“ˆ Best mAP@50: {results.box.map50:.4f}")

# ì‹¤í–‰
if __name__ == "__main__":
    # Baseline ëª¨ë¸ë¡œ YOLOv8 Nano ì„ ì •
    run_baseline_training("yolov8n", device, data_yaml, save_dir)
```
