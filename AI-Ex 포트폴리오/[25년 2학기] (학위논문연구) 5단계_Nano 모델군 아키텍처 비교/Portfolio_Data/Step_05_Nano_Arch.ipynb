```python

from ultralytics import YOLO
import os

# í™˜ê²½ ì„¤ì •
base_path = r"D:\Virtual_Fahion\GraduationThesis2"
data_yaml = os.path.join(base_path, "data", "data.yaml")
save_dir = os.path.join(base_path, "saveModels")
device = '0'

def run_nano_comparison(device, data_path, save_path):
    print("\nğŸš€ [Step 5] Starting Nano Model Architecture Comparison")
    
    # Nano ëª¨ë¸ ë¦¬ìŠ¤íŠ¸ (v5, v8, v10)
    nano_models = ['yolov5n', 'yolov8n', 'yolov10n']
    comparison_results = {}
    
    for m_name in nano_models:
        print(f"\nâ–¶ï¸ Training Architecture: {m_name}")
        try:
            # v5ëŠ” yolov5nu.pt ì‚¬ìš© ê¶Œì¥
            weight_file = f"{m_name}u.pt" if 'v5' in m_name else f"{m_name}.pt"
            model = YOLO(weight_file)
            
            results = model.train(
                data=data_path, epochs=30, imgsz=640, device=device,
                project=save_path, name=f"arch_{m_name}", exist_ok=True,
                augment=True, verbose=False
            )
            comparison_results[m_name] = results.box.map50
        except Exception as e:
            print(f"âš ï¸ Failed to train {m_name}: {e}")
            
    print("\nğŸ† Architecture Comparison Results (mAP@50):")
    for m, score in comparison_results.items():
        print(f" - {m}: {score:.4f}")

# ì‹¤í–‰
if __name__ == "__main__":
    run_nano_comparison(device, data_yaml, save_dir)
```
