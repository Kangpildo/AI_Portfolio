```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# (Step 7ì˜ í•¨ìˆ˜ê°€ í•„ìš”í•˜ë¯€ë¡œ ì½”ë“œë¥¼ í•©ì¹˜ê±°ë‚˜ ëª¨ë“ˆë¡œ ë¶ˆëŸ¬ì™€ì•¼ í•¨)
# ì—¬ê¸°ì„œëŠ” ê²°ê³¼ ë°ì´í„°ê°€ ìˆë‹¤ê³  ê°€ì •í•˜ê³  ì‹œê°í™”ë§Œ ìˆ˜í–‰í•˜ëŠ” ì½”ë“œ ì˜ˆì‹œì…ë‹ˆë‹¤.

# ì˜ˆì‹œ ë°ì´í„° (ì‹¤ì œë¡œëŠ” Step 7ì—ì„œ ë°˜í™˜ëœ ë°ì´í„°ë¥¼ ì‚¬ìš©)
data = [
    {'Model': 'yolov5n', 'mAP50': 0.65, 'Latency(ms)': 1.8, 'Params(M)': 2.5},
    {'Model': 'yolov8n', 'mAP50': 0.68, 'Latency(ms)': 2.0, 'Params(M)': 3.2},
    {'Model': 'yolov10n', 'mAP50': 0.67, 'Latency(ms)': 1.5, 'Params(M)': 2.3},
    {'Model': 'yolov8s', 'mAP50': 0.72, 'Latency(ms)': 4.5, 'Params(M)': 11.1}
]

def final_analysis(data_list):
    print("\nğŸš€ [Step 8] Final Trade-off Analysis")
    
    df = pd.DataFrame(data_list)
    df_sorted = df.sort_values(by='mAP50', ascending=False)
    
    print("\nğŸ† Final Leaderboard:")
    print(df_sorted.to_string(index=False))
    
    # Trade-off ì‹œê°í™”
    plt.figure(figsize=(10, 6))
    sns.set_style("whitegrid")
    
    # X: ì†ë„, Y: ì •í™•ë„, í¬ê¸°: íŒŒë¼ë¯¸í„° ìˆ˜
    sns.scatterplot(
        data=df, x="Latency(ms)", y="mAP50", 
        hue="Model", style="Model", size="Params(M)", sizes=(200, 1000), 
        palette="viridis", alpha=0.8
    )
    
    for i in range(df.shape[0]):
        plt.text(
            df.loc[i, "Latency(ms)"]+0.1, df.loc[i, "mAP50"], 
            df.loc[i, "Model"], fontdict={'size':10, 'weight':'bold'}
        )
        
    plt.title("Accuracy vs Speed Trade-off Analysis", fontsize=16)
    plt.xlabel("Latency (ms) [Lower is Faster]", fontsize=12)
    plt.ylabel("mAP@50 [Higher is Better]", fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    
    plt.savefig("final_tradeoff_analysis.png")
    plt.show()
    print(f"\nğŸ–¼ï¸ Analysis Plot saved as 'final_tradeoff_analysis.png'")

# ì‹¤í–‰
if __name__ == "__main__":
    # ì‹¤ì œë¡œëŠ” Step 7ì˜ extract_metrics í•¨ìˆ˜ ê²°ê³¼ë¥¼ ë„£ì–´ì£¼ì„¸ìš”
    final_analysis(data)
```
