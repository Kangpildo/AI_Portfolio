```python

from ultralytics import YOLO
import os

# í™˜ê²½ ì„¤ì •
base_path = r"D:\Virtual_Fahion\GraduationThesis2"
data_yaml = os.path.join(base_path, "data", "data.yaml")
save_dir = os.path.join(base_path, "saveModels")
device = '0'

def run_scaling_experiment(device, data_path, save_path):
    print("\nğŸš€ [Step 6] Starting Model Scaling Experiment (Nano vs Small)")
    
    # v8n(Nano) vs v8s(Small) ë¹„êµ
    target_pairs = [('yolov8n', 'yolov8s')] 
    
    for nano, small in target_pairs:
        print(f"\nâ–¶ï¸ Comparing {nano} vs {small}")
        
        # Small ëª¨ë¸ í•™ìŠµ (NanoëŠ” ì´ì „ ë‹¨ê³„ ê²°ê³¼ í™œìš© ê°€ëŠ¥í•˜ë‚˜ ì—¬ê¸°ì„  íë¦„ìƒ ìƒëµí•˜ê±°ë‚˜ ì¬í•™ìŠµ)
        model_s = YOLO(f"{small}.pt")
        results_s = model_s.train(
            data=data_path, epochs=30, imgsz=640, device=device,
            project=save_path, name=f"scale_{small}", exist_ok=True,
            augment=True, verbose=False
        )
        
        print(f"ğŸ“Š Scaling Results for {small}:")
        print(f" - mAP@50: {results_s.box.map50:.4f}")

# ì‹¤í–‰
if __name__ == "__main__":
    run_scaling_experiment(device, data_yaml, save_dir)
```
