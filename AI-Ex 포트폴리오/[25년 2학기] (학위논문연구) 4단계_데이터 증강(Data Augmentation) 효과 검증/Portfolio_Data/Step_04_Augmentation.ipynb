```python

from ultralytics import YOLO
import os

# í™˜ê²½ ì„¤ì •
base_path = r"D:\Virtual_Fahion\GraduationThesis2"
data_yaml = os.path.join(base_path, "data", "data.yaml")
save_dir = os.path.join(base_path, "saveModels")
device = '0'

def run_augmentation_experiment(model_name, device, data_path, save_path):
    print("\nğŸš€ [Step 4] Starting Augmentation Ablation Study")
    
    # ë¹„êµêµ° ì„¤ì •: ì¦ê°• ë„ê¸°(False) vs ì¼œê¸°(True)
    augment_settings = [False, True]
    results_summary = {}
    
    for aug_flag in augment_settings:
        run_name = f"exp_{model_name}_aug_{aug_flag}"
        print(f"\nâ–¶ï¸ Running Experiment: Augmentation = {aug_flag}")
        
        model = YOLO(f"{model_name}.pt")
        results = model.train(
            data=data_path, epochs=30, imgsz=640, device=device,
            project=save_path, name=run_name, exist_ok=True,
            augment=aug_flag, # [í•µì‹¬] ë°ì´í„° ì¦ê°• On/Off ì œì–´
            verbose=False
        )
        results_summary[f"Augment_{aug_flag}"] = results.box.map50
        
    print("\nğŸ“Š Ablation Study Results (mAP@50):")
    for setting, score in results_summary.items():
        print(f" - {setting}: {score:.4f}")

# ì‹¤í–‰
if __name__ == "__main__":
    run_augmentation_experiment("yolov8n", device, data_yaml, save_dir)
```
